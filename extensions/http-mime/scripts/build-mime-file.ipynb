```python
import json

cat_map = {}
category = None

def ensure_cat(new_cat:str):
  global category
  global cat_map
  if category != new_cat and category is not None:
    print(cat_map[category])
  category = new_cat
  cat_map[category] = cat_map[category] if category in cat_map else {}

def add_mime_type(mime:str, exts:list, description:str=None):
  global category
  global cat_map
  cat = mime.split('/')[0]
  ensure_cat(cat)
  if mime not in cat_map[cat]:
    cat_map[cat][mime] = {
      'exts':exts,
      'description':description
    }
  else:
    cat_map[cat][mime]['exts'].extend(exts)
    cat_map[cat][mime]['exts'] = list(set(cat_map[cat][mime]['exts']))
    if description:
      cat_map[cat][mime]['description'] = description



with open('./Apache_Configuration_htaccess') as f:
  for l in f.readlines():
    if l.startswith('<'):
      continue
    if l.strip().startswith('#'):
      continue
    else:
      items = l.split()
      mime = items[1]
      exts = [f'.{ext}' for ext in items[2:]]
      add_mime_type(mime,exts)


from bs4 import BeautifulSoup
with open('./common-mime-types.html') as f:
  soup = BeautifulSoup(f, 'html.parser')
  for row in soup.table.tbody.findAll('tr'):
     col1 = row.td
     exts = col1.code.contents
     col2 = col1.findNextSibling()
     description = col2.get_text()
     mime = col2.findNextSibling().code.get_text()
    #  print('add',mime,exts,description)
     add_mime_type(mime,exts,description)

with open('../src/mime-map.json','w') as f:
  json.dump(cat_map,f)

```
